```python
# Importing Python libraries
import numpy as np
import pandas as pd
```

```python
# Creating a dataframe for Japan 

# Importing from github
df1 = pd.read_excel('https://github.com/alexwonglse/EC1B1-Group-45-Japan-Data/blob/main/International_Financial_Statistics_Japan.xlsx?raw=true')

# cleaning the data
df1.drop(0,axis=0,inplace=True)

# Renaming the columns 
df1 = df1.rename(columns={'International Financial Statistics (IFS)': 'Month/Year'})
df1 = df1.rename(columns={'Unnamed: 1': 'Economic Activity, Industrial Production, Index'})
df1 = df1.rename(columns={'Unnamed: 2': 'Exchange Rates, National Currency Per U.S. Dollar, Period Average, Rate'})
df1 = df1.rename(columns={'Unnamed: 3': 'International Reserves and Liquidity, Reserves, Official Reserve Assets, US Dollar, Japan'})
df1 = df1.rename(columns={'Unnamed: 4': 'Prices, Consumer Price Index, All items, Index, Japan'})
df1.head(2)

df1
```

```python
# Creating a dataframe for US

# Importing data from github
df2 = pd.read_excel('https://github.com/alexwonglse/EC1B1-Group-45-US-Data/blob/main/International_Financial_Statistics_US.xlsx?raw=true')

# Cleaning the data
df2.drop(0,axis=0,inplace=True)

# Renaming the columns 
df2 = df2.rename(columns={'International Financial Statistics (IFS)': 'Month/Year'})
df2 = df2.rename(columns={'Unnamed: 1': 'International Reserves and Liquidity, Reserves, Official Reserve Assets, US Dollar, US'})
df2 = df2.rename(columns={'Unnamed: 2': 'Prices, Consumer Price Index, All items, Index, US'})


df2
```

```python
# Merge Datasets 
import pandas as pd 
df3 = pd.merge(df1, df2, on = 'Month/Year')
df3.head(3)
```

```python
# Checking for Missing Values (NaN) in the Dataframe
df3.isnull().values.any()
```

```python
# Calculating Japan Log Exchange Rate

# Turning Exchange Rates, National Currency Per U.S. Dollar, Period Average, Rate into a float 
df3['Exchange Rates, National Currency Per U.S. Dollar, Period Average, Rate'] = df3['Exchange Rates, National Currency Per U.S. Dollar, Period Average, Rate'].astype(float)

# Taking the natural log of the Exchange Rates, National Currency Per U.S. Dollar, Period Average, Rate column
df3['Japan: Log Exchange Rate'] = np.log(df3['Exchange Rates, National Currency Per U.S. Dollar, Period Average, Rate'])
df3
```

```python
# Calculating the difference in the Log Exchange Rate versus the previous month for Japan
df3['Japan: Log Exchange Rate Difference'] = df3['Japan: Log Exchange Rate'].diff()
df3.head(3)
```

```python
# Calculating the Monthly Inflation Rate for Japan
df3['Japan: Monthly Inflation Rate'] = df3['Prices, Consumer Price Index, All items, Index, Japan'].pct_change()
df3.head(3)
```

```python
# Calculating the Log Industrial Production for Japan

# Turning Economic Activity, Industrial Production, Index into a float
df3['Economic Activity, Industrial Production, Index'] = df3['Economic Activity, Industrial Production, Index'].astype(float)

# Taking natural log of the Economic Activity, Industrial Production, Index column
df3['Japan: Log Industrial Production'] = np.log(df3['Economic Activity, Industrial Production, Index'])
df3.head(3)
```

```python
# Calculating Monthly Growth in Industrial Production for Japan
df3['Japan: Monthly Growth in Industrial Production'] = df3['Japan: Log Industrial Production'].diff()
df3.head(3)
```

```python
# Calculating the Growth in Industrial Production versus 12 months ago for Japan
df3['Growth in industrial production versus 12 months ago'] = df3['Japan: Log Industrial Production'].pct_change(periods=12)
df3.head(3)
```

```python
# Indexing in Log Changes for International Reserves for Japan

# Using the formula: index = ln(row) - ln(row at Japan 1960)

index_japan_dummy_variable = df3['International Reserves and Liquidity, Reserves, Official Reserve Assets, US Dollar, Japan'] - df3.loc[0,'International Reserves and Liquidity, Reserves, Official Reserve Assets, US Dollar, Japan']
df3['Index of the value of international reserves, Japan'] = 100+(index_japan_dummy_variable / df3.loc[0,'International Reserves and Liquidity, Reserves, Official Reserve Assets, US Dollar, Japan'])
df3.head(3)
```

```python
# Calculating the Monthly Inflation Rate for US
df3['US: Monthly Inflation Rate'] = df3['Prices, Consumer Price Index, All items, Index, US'].pct_change()
df3.head(3)
```

```python
# Indexing in Log Changes for International Reserves for US
# Using the formula: index = ln(row) - ln(row at jan 1960)
import pandas as pd
import numpy as np

index_US_dummy_variable = df3['International Reserves and Liquidity, Reserves, Official Reserve Assets, US Dollar, US'] - df3.loc[0,'International Reserves and Liquidity, Reserves, Official Reserve Assets, US Dollar, US']
df3['Index of the value of international reserves, US'] = 100+(index_US_dummy_variable / df3.loc[0,'International Reserves and Liquidity, Reserves, Official Reserve Assets, US Dollar, US'])
df3.head(3)
```

```python
# Getting rid of outliers 

from scipy import stats
import numpy as np

def replace_outliers_with_mean(series):

#Replace outliers in a pandas series with the mean of the previous and next item.

    outlier_indices = np.where(np.abs(stats.zscore(series)) > 3)[0]
    for i in outlier_indices:
          series[i] = (series[i-1] + series[i+1]) / 2
    return series

for col in df3.columns:
    if df3[col].dtype == np.number:
        z = np.abs(stats.zscore(df3[col]))
        outliers = df3[col][z > 3]
        if len(outliers) < 5:
            df3[col] = replace_outliers_with_mean(df3[col])
        else: 
          continue
df3.head(3) 
```

```python
# Using the following formula: Real Exchange Growth Rate = (Growth in Nominal Exchange Rate + Domestic Currency per Dollar-Price Index, for the US)
ex_rate_growth = df3['Japan: Log Exchange Rate Difference']
us_inflation = df3['US: Monthly Inflation Rate']
japan_inflation = df3['Japan: Monthly Inflation Rate']
df3['Monthly Real Exchange Exchange-Rate Growth'] = ex_rate_growth-japan_inflation+us_inflation
df3.head(3)
```

```python
# Convert the Month/Year column to a Datetime format
date_format = pd.to_datetime(df3['Month/Year'], format='%b %Y')
df3.insert(0, 'Date', date_format)
df3
```

```python
# Divide DataFrame by 10 months from date '1973-02-01'
df_pre = df3.loc[df3['Date'] <= '1972-05-01']
df_post = df3.loc[df3['Date'] > '1973-12-01']
```

```python
# Calculate Standard Deviation for each variable
pre_std_ex_growth_nom = df_pre['Japan: Log Exchange Rate Difference'].std()
post_std_ex_growth_nom = df_post['Japan: Log Exchange Rate Difference'].std()
pre_std_ex_growth_real = df_pre['Monthly Real Exchange Exchange-Rate Growth'].std()
post_std_ex_growth_real = df_post['Monthly Real Exchange Exchange-Rate Growth'].std()
pre_inflation = df_pre['Japan: Monthly Inflation Rate'].std()
post_inflation = df_post['Japan: Monthly Inflation Rate'].std()
pre_diff_inflation = (df_pre['Japan: Monthly Inflation Rate'] - df_pre['US: Monthly Inflation Rate']).std()
post_diff_inflation = (df_post['Japan: Monthly Inflation Rate'] - df_post['US: Monthly Inflation Rate']).std()
pre_ip_growth = df_pre['Growth in industrial production versus 12 months ago'].std()
post_ip_growth = df_post['Growth in industrial production versus 12 months ago'].std()
```

```python
# Create a table for the data regarding post and pre Bretton Woods
data = {'Variable': ['Log exchange rate difference', 'Real exchange rate growth', 'Inflation rate', 'Inflation difference', 'IP growth'], 
        'Pre Bretton Woods Exit': [pre_std_ex_growth_nom, pre_std_ex_growth_real, pre_inflation, pre_diff_inflation, pre_ip_growth], 
        'Post Bretton Woods Exit': [post_std_ex_growth_nom, post_std_ex_growth_real, post_inflation, post_diff_inflation, post_ip_growth]}  
table_df = pd.DataFrame(data)
table_df.set_index('Variable', inplace = True)
```

```python
# Print the table for the data regarding Bretton Woods
table_df
```

```python
# Calculate ratio of before and after for each variable 
ratio_logexrate = pre_std_ex_growth_nom / post_std_ex_growth_nom
ratio_realexrate = pre_std_ex_growth_real / post_std_ex_growth_real
ratio_inflation = pre_inflation / post_inflation
ratio_diff_inflation = pre_diff_inflation / post_diff_inflation
ratio_ip_growth = pre_ip_growth / post_ip_growth
```
